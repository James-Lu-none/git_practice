```python
# hjson load
import json
import numpy as np
import os
import pickle
import random
import string
from pathlib import Path
from data_set_trans import *

current_file_path = Path(get_ipython().run_line_magic('pwd', ''))
root_path = current_file_path
while not any(file.suffix == ".ipynb" for file in root_path.glob("*")):
    root_path = root_path.parent

root_path = str(root_path)
print(root_path)

data_dir = os.path.join(root_path,'json_manual_data')


def rewrite_game_from_series_move(moves):
    board = np.zeros((15, 15))
    board -= 1
    for idx, i in enumerate(moves):
        board[int(i/15), i % 15] = idx
    # print(board)
    return board


json_data_files = os.listdir(data_dir)
all_data = []
for idx, f in enumerate(json_data_files):
    print("loading file {}...".format(idx))
    f = os.path.join(data_dir, f)
    with open(f, 'rb') as f:
        file_data = json.load(f)
    f.close()
    print("file {} loaded".format(idx))
    gameKeys = file_data.keys()
    for i in gameKeys:
        seriesdata = file_data[i]['moves']
        all_data.append(rewrite_game_from_series_move(seriesdata))

for i in [1,2,3,4]:
    last_n=i
    # -------------------------------
    data_final = []
    for game_state in np.array(data_augmentation(all_data)):
        data_final.append(data_trans(game_state,last_n))
    # -------------------------------
    save_trans_data(root_path,data_final,'data_from_json',last_n)


# for i in data_final[0][2][0]:
#     print(i)

# #[第幾盤][第幾手][0]  [特徵0~3][x][y] 局勢
# #[第幾盤][第幾手][1] 機率
# #[第幾盤][第幾手][2] 誰贏
```
